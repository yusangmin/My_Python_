# 대화형 프로그래밍
 ## REPL(Read-Eval-Print Loop)
 ### 3. 객체(object)
  - 명령형(imperative)프로그래밍
    - 절차적 프로그래밍
    - 객체 지향적 프로그래밍

 - 객체 지향 프로그래밍이란
    - 데이터와 코드가 결합된 객체를 기반으로 만드는 프로그래밍
    - C++, 파이썬, 자마(<--> C, 포트란)
 - 파이썬은 완전한 객체 지향 프로그래밍 언어
    - 절차적 특징도 가진 멀티-패러다임 언어

 - 객체의 종류
   - 스칼라(scalar)
      - 정수, 실수(부동소수점 숫자), 불리언(boolean), None
      - None은 값이 업음을 나타내는 객체로 0이 아니다.
   - 비스칼라
      - 문자열, 튜플, 집합, 리스트, 딕셔너리

 - 표현식
   - 객체와 연산자(operator)의 연결
     - 3 + 4
     - 2.5 * 2
     - True and False
     - type(3)

   - 표현식의 결과는 값
     - 7
     - 5.0
     - False
     - int
    
- 변수: 값을 저장(?) 또는 할당(?)
  - 변수는 객체에 이름을 부여한다.
    - pi = 3
    - id(pi), id(3)
